```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("gas_prices.csv",delimiter=",")
```

```python
data.head()
```

```python
plt.plot(data["Year"],data["Australia"],".--",label="Australia")
plt.plot(data["Year"],data["USA"],".--",label="USA")
plt.plot(data["Year"],data["France"],".--",label="France")
plt.legend()
plt.xticks(data["Year"][::2].to_list()+[2010])
plt.xlabel("Years")
plt.ylabel("Price in USD")
plt.title("Inflation In Gas Prices")
plt.savefig("Gas Inflation",format="png",dpi=500)
plt.show()
```

```python
plt.figure(figsize=[7,5],linewidth=5)

for country in data.columns[1:]:
    plt.plot(data["Year"],data[country],".--",label=country)


plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(data["Year"][::2].to_list()+[2010])
plt.xlabel("Years")
plt.ylabel("Price in USD")
plt.title("Inflation In Gas Prices")
plt.savefig("Gas Inflation In Different Countries",format="png",dpi=500,bbox_inches='tight')
plt.show()
```
